#1- Quantos óbitos, registrados em internações no SUS, ocorreram de janeiro de 2020 a setembro de 2025, em relação a todos os óbitos ocorridos, registrados em internações no SUS? 

```sql
SELECT 
    SUM(`2020`) as `2020`,
    SUM(`2021`) as `2021`,
    SUM(`2022`) as `2022`,
    SUM(`2023`) as `2023`,
    SUM(`2024`) as `2024`,
    SUM(`2025`) as `2025`,
    
   SUM(`2020`) + SUM(`2021`) + SUM(`2022`) + SUM(`2023`) + SUM(`2024`) + SUM(`2025`) AS Total_obitos
FROM 
    obitos_geral;
```

#Qual Unidade da Federação apresentou o maior total de óbitos por HIV provenientes de internações no SUS no período de janeiro de 2020 a setembro de 2025 e quantos foram esses óbitos?

```sql
create or replace view mvp_datasus.silver.obitos_hiv_maior AS
SELECT
    unidade_federacao as Estado,
    (`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) AS Total
FROM obitos_hiv_geral
ORDER BY Total desc
LIMIT 1;
```

#4- Qual é a distribuição dos óbitos por HIV, por faixa etária, entre os registrados a partir de internações no SUS no período de janeiro de 2020 a setembro de 2025?

```sql
use mvp_datasus.silver
```

```sql
CREATE OR REPLACE VIEW mvp_datasus.silver.obitos_hiv_faixas AS
SELECT
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_0_10)   AS `Faixa 0-10 anos`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_10_25)  AS `Faixa 10-25 anos`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_25_40)  AS `Faixa 25-40 anos`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_40_55)  AS `Faixa 40-55 anos`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_55_65)  AS `Faixa 55-65 anos`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_faixa_65_mais) AS `Faixa 65+ anos`;
```

```sql
select * from mvp_datasus.silver.obitos_hiv_faixas
```

#5- Qual faixa etária concentrou o maior número de óbitos por HIV decorrentes de internações no SUS entre janeiro de 2020 e setembro de 2025 e quantos foram esses óbitos?

```sql
CREATE OR REPLACE VIEW obitos_hiv_faixa_maior AS
SELECT
    GREATEST(
        `Faixa 0-10 anos`,
        `Faixa 10-25 anos`,
        `Faixa 25-40 anos`,
        `Faixa 40-55 anos`,
        `Faixa 55-65 anos`,
        `Faixa 65+ anos`
    ) AS maior_valor,
    CASE GREATEST(
        `Faixa 0-10 anos`,
        `Faixa 10-25 anos`,
        `Faixa 25-40 anos`,
        `Faixa 40-55 anos`,
        `Faixa 55-65 anos`,
        `Faixa 65+ anos`
    )
    WHEN `Faixa 0-10 anos`   THEN 'Faixa 0-10 anos'
    WHEN `Faixa 10-25 anos`  THEN 'Faixa 10-25 anos'
    WHEN `Faixa 25-40 anos`  THEN 'Faixa 25-40 anos'
    WHEN `Faixa 40-55 anos`  THEN 'Faixa 40-55 anos'
    WHEN `Faixa 55-65 anos`  THEN 'Faixa 55-65 anos'
    WHEN `Faixa 65+ anos`    THEN 'Faixa 65+ anos'
    END AS faixa_com_maior_valor
FROM mvp_datasus.silver.obitos_hiv_faixas;
```

```sql
select * from obitos_hiv_faixa_maior
```

#6- Qual faixa etária concentrou o menor número de óbitos por HIV decorrentes de internações no SUS entre janeiro de 2020 e setembro de 2025 e quantos foram esses óbitos?

```sql
CREATE OR REPLACE VIEW obitos_hiv_faixa_menor AS
SELECT
    LEAST(
        `Faixa 0-10 anos`,
        `Faixa 10-25 anos`,
        `Faixa 25-40 anos`,
        `Faixa 40-55 anos`,
        `Faixa 55-65 anos`,
        `Faixa 65+ anos`
    ) AS menor_valor,
    CASE LEAST(
        `Faixa 0-10 anos`,
        `Faixa 10-25 anos`,
        `Faixa 25-40 anos`,
        `Faixa 40-55 anos`,
        `Faixa 55-65 anos`,
        `Faixa 65+ anos`
    )
    WHEN `Faixa 0-10 anos`   THEN 'Faixa 0-10 anos'
    WHEN `Faixa 10-25 anos`  THEN 'Faixa 10-25 anos'
    WHEN `Faixa 25-40 anos`  THEN 'Faixa 25-40 anos'
    WHEN `Faixa 40-55 anos`  THEN 'Faixa 40-55 anos'
    WHEN `Faixa 55-65 anos`  THEN 'Faixa 55-65 anos'
    WHEN `Faixa 65+ anos`    THEN 'Faixa 65+ anos'
    END AS faixa_com_menor_valor
FROM mvp_datasus.silver.obitos_hiv_faixas;
```

```sql
SELECT * FROM obitos_hiv_faixa_menor;
```

#7- Como se distribuem, por sexo, os óbitos por HIV decorrentes de internações no SUS no período de janeiro de 2020 a setembro de 2025?

```sql
CREATE OR REPLACE VIEW mvp_datasus.silver.obitos_hiv_sexo AS
SELECT
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_masc)   AS `Masculino`,
    (SELECT SUM(`2020` + `2021` + `2022` + `2023` + `2024` + `2025`) FROM mvp_datasus.silver.obitos_hiv_fem)  AS `Feminino`;
```

```sql
select * from mvp_datasus.silver.obitos_hiv_sexo
```
